```sql
SELECT 
transaction_date,
dayname(transaction_date) AS Day_of_Transaction,
monthname(transaction_date) AS Month_of_Transaction,
--date_format(transaction_time, 'HH:mm:ss') AS Transaction_Time,
dayofmonth(transaction_date) AS Actual_Day_of_Month,

case 
when Day_of_Transaction in ('Sun', 'Sat') THEN 'Weekend'
ELSE 'Weekday'
END AS Day_Category,

case 
when date_format(transaction_time, 'HH:mm:ss') BETWEEN '00:00:00' AND '11:59:59' THEN 'Morning'
when date_format(transaction_time, 'HH:mm:ss') BETWEEN '12:00:00' AND '17:59:59' THEN 'Afternoon'
when date_format(transaction_time, 'HH:mm:ss') >= '18:00:00' THEN 'Evening'
END AS Time_Bucket,


COUNT(DISTINCT transaction_id) AS Total_Sales_Made,
COUNT(DISTINCT product_id) AS Total_Products,
COUNT(DISTINCT store_id) AS Total_Stores,


SUM(transaction_qty*unit_price) AS Sale_Per_Transaction,

Case 
When Sale_Per_Transaction <= 50 THEN 'Low Spend'
When Sale_Per_Transaction Between 51 and 100 Then 'Medium Spend'
ELSE 'High Spend'
END AS Spend_Bucket,

store_location,
product_category,
product_detail

FROM bright_coffee_shop_analysis_case_study_1

GROUP BY 
transaction_date,
Day_of_Transaction,
store_location,
product_category,
product_detail,
--Transaction_Time,
Time_Bucket,
Actual_Day_of_Month,
Month_of_Transaction;
```
